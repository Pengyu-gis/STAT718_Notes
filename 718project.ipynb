```python
import pandas as pd
from sklearn.neighbors import NearestNeighbors

# Load accident data (filtered where NumberOfInjuries >= 1)
accidents = pd.read_csv('accidents_with_injuries.csv')

# Load slope & curve data
slope_curve = pd.read_csv('/content/drive/MyDrive/stat718/slope_curve_output.csv')

# Prepare coordinates
accident_coords = accidents[['X', 'Y']].values
slope_coords = slope_curve[['x', 'y']].values

# Fit KNN (using 1 nearest neighbor)
knn = NearestNeighbors(n_neighbors=1, algorithm='auto')
knn.fit(slope_coords)

# Find the nearest slope/curve point for each accident
distances, indices = knn.kneighbors(accident_coords)

# Assign slope and curve values
accidents['slope'] = slope_curve.iloc[indices.flatten()]['value'].values
accidents['curve'] = slope_curve.iloc[indices.flatten()]['r'].values

# Save the enriched dataset
accidents.to_csv('accidents_with_slope_curve.csv', index=False)

print("✅ Slope and curve matched and saved!")
```

# point process

```python
import geopandas as gpd
import pandas as pd
from shapely.geometry import Point

# Load cleaned dataset
df = pd.read_csv('accidents_cleaned.csv')

# Convert to GeoDataFrame
# Create GeoDataFrame from X, Y
geometry = [Point(xy) for xy in zip(df['X'], df['Y'])]
gdf = gpd.GeoDataFrame(df, geometry=geometry, crs="EPSG:4326")
```

```python
covariates = pd.get_dummies(
    gdf[['LightingCondition', 'WeatherCondition', 'RoadSurfaceCondition']],
    drop_first=True,
    dtype=int
)


covariates = covariates.rename(columns={
    'slope': 'slope_cov',
    'curve': 'curve_cov',
    'hour': 'hour_cov'
})

# Join back with spatial data, specifying suffixes for overlapping columns
gdf = gdf.join(covariates, lsuffix='_gdf', rsuffix='_cov')

```

```python
import statsmodels.api as sm

# Drop rows with missing values
model_data = covariates.dropna()

# Add intercept
model_data = sm.add_constant(model_data)

# Target: one count per accident point
counts = [1] * len(model_data)

# Fit Poisson regression model
poisson_model = sm.GLM(counts, model_data, family=sm.families.Poisson()).fit()

# Show the result
print(poisson_model.summary())
```

# build grid

```python
import geopandas as gpd
import pandas as pd
import numpy as np
from shapely.geometry import Point, box

# Load accident GeoDataFrame
accidents = pd.read_csv('accidents_cleaned.csv')
accidents_gdf = gpd.GeoDataFrame(accidents, geometry=gpd.points_from_xy(accidents.X, accidents.Y), crs="EPSG:4326")

# Reproject to meters for better grid spacing
accidents_gdf = accidents_gdf.to_crs(epsg=3857)

# Bounding box of all accidents
minx, miny, maxx, maxy = accidents_gdf.total_bounds

# Grid size (e.g., 5 km x 5 km)
cell_size = 5000  # meters

# Create fishnet/grid
grid_cells = []
for x in np.arange(minx, maxx, cell_size):
    for y in np.arange(miny, maxy, cell_size):
        grid_cells.append(box(x, y, x + cell_size, y + cell_size))

grid = gpd.GeoDataFrame({'geometry': grid_cells}, crs="EPSG:3857")
```

```python
# Spatial join: assign accidents to grid cells
joined = gpd.sjoin(accidents_gdf, grid, how="left", predicate="within")

# Count accidents per cell
counts_per_cell = joined.groupby('index_right').size().reset_index(name='accident_count')

# Merge back into grid
grid['accident_count'] = 0
grid.loc[counts_per_cell['index_right'], 'accident_count'] = counts_per_cell['accident_count'].values
```

```python
# Average slope and curve
agg_slope_curve = joined.groupby('index_right')[['slope', 'curve']].mean().reset_index()

# Merge into grid
grid.loc[agg_slope_curve['index_right'], 'slope'] = agg_slope_curve['slope'].values
grid.loc[agg_slope_curve['index_right'], 'curve'] = agg_slope_curve['curve'].values
```

```python
import statsmodels.api as sm

# Drop empty or NaN cells
grid_model = grid.dropna(subset=['accident_count', 'slope', 'curve'])

# Features and target
X = grid_model[['slope', 'curve']]
X = sm.add_constant(X)
y = grid_model['accident_count']

# Fit model
poisson_model = sm.GLM(y, X, family=sm.families.Poisson()).fit()
print(poisson_model.summary())
```

# add more covariates

```python
# -----------------------------
# Step 1: One-Hot Encode Categorical Covariates
# -----------------------------
# Convert categorical variables into dummies with 0/1 values.
# (Drop the first category to avoid collinearity.)
dummies = pd.get_dummies(
    accidents_gdf[['LightingCondition', 'WeatherCondition', 'RoadSurfaceCondition']],
    drop_first=True,
    dtype=int
)
```

```python
# Join the dummy variables back into the GeoDataFrame
accidents_gdf = accidents_gdf.join(dummies)

# -----------------------------
# Step 2: Create a Background Grid (Fishnet)
# -----------------------------
# Get the bounding box of the accident data
minx, miny, maxx, maxy = accidents_gdf.total_bounds

# Define grid cell size (e.g., 5 km x 5 km)
cell_size = 5000  # in meters

# Build grid cells covering the study area
grid_cells = []
for x in np.arange(minx, maxx, cell_size):
    for y in np.arange(miny, maxy, cell_size):
        grid_cells.append(box(x, y, x + cell_size, y + cell_size))

grid = gpd.GeoDataFrame({'geometry': grid_cells}, crs="EPSG:3857")
```

```python
# -----------------------------
# Step 3: Spatial Join & Aggregation
# -----------------------------
# Spatially join accidents to grid cells (each accident gets the index of the cell it falls into)
joined = gpd.sjoin(accidents_gdf, grid, how="left", predicate="within")

# 3a. Count the number of accidents in each grid cell
counts_per_cell = joined.groupby('index_right').size().reset_index(name='accident_count')
grid['accident_count'] = 0
grid.loc[counts_per_cell['index_right'], 'accident_count'] = counts_per_cell['accident_count'].values

# 3b. Compute average slope and curve per grid cell
agg_slope_curve = joined.groupby('index_right')[['slope', 'curve']].mean().reset_index()
grid.loc[agg_slope_curve['index_right'], 'slope'] = agg_slope_curve['slope'].values
grid.loc[agg_slope_curve['index_right'], 'curve'] = agg_slope_curve['curve'].values

# 3c. Aggregate dummy variables (proportions of each condition) per grid cell.
dummy_cols = dummies.columns.tolist()
agg_dummies = joined.groupby('index_right')[dummy_cols].mean().reset_index()
for col in dummy_cols:
    grid.loc[agg_dummies['index_right'], col] = agg_dummies[col].values
```

```python
# -----------------------------
# Step 4: Fit the Poisson Regression Model
# -----------------------------
# Drop grid cells that have missing covariate values (or accident_count is NaN)
grid_model = grid.dropna(subset=['accident_count', 'slope', 'curve'])

# Select features for the model:
# Here, we use slope, curve, and the aggregated dummy covariates.
model_features = ['slope', 'curve'] + dummy_cols
X = grid_model[model_features]

# Add an intercept
X = sm.add_constant(X)

# The response is the accident count per grid cell.
y = grid_model['accident_count']

# Fit the Poisson GLM
poisson_model = sm.GLM(y, X, family=sm.families.Poisson()).fit()

print(poisson_model.summary())
```

# add temporal dimension

```python
# -----------------------------
# Step 1: Load and Prepare Accident Data
# -----------------------------
# Load your cleaned accident dataset
accidents = pd.read_csv('accidents_cleaned.csv')

# Create GeoDataFrame
accidents_gdf = gpd.GeoDataFrame(
    accidents,
    geometry=gpd.points_from_xy(accidents.X, accidents.Y),
    crs="EPSG:4326"
)

# Reproject to meters for spatial grid
accidents_gdf = accidents_gdf.to_crs(epsg=3857)
```

```python
def categorize_hour(hour):
    if 0 <= hour <= 5:
        return 'wee_hours'
    elif 6 <= hour <= 10:
        return 'morning'
    elif 11 <= hour <= 15:
        return 'noon'
    elif 16 <= hour <= 20:
        return 'night'
    else:
        return 'midnight'

# Apply to create new time-of-day column
accidents_gdf['time_of_day'] = accidents_gdf['hour'].apply(categorize_hour)
```

```python
# -----------------------------
# Step 3: One-Hot Encode All Categorical Variables
# -----------------------------
# Lighting, weather, surface condition
dummies1 = pd.get_dummies(
    accidents_gdf[['LightingCondition', 'WeatherCondition', 'RoadSurfaceCondition']],
    drop_first=True, dtype=int
)

# Time of day
dummies2 = pd.get_dummies(accidents_gdf['time_of_day'], drop_first=True, dtype=int)

# Combine all dummy variables
accidents_gdf = accidents_gdf.join(dummies1)
accidents_gdf = accidents_gdf.join(dummies2)

# Collect dummy column names
dummy_cols = dummies1.columns.tolist()
time_dummy_cols = dummies2.columns.tolist()
all_dummy_cols = dummy_cols + time_dummy_cols
```

```python
# -----------------------------
# Step 4: Create Background Grid
# -----------------------------
minx, miny, maxx, maxy = accidents_gdf.total_bounds
cell_size = 5000  # meters

grid_cells = []
for x in np.arange(minx, maxx, cell_size):
    for y in np.arange(miny, maxy, cell_size):
        grid_cells.append(box(x, y, x + cell_size, y + cell_size))

grid = gpd.GeoDataFrame({'geometry': grid_cells}, crs="EPSG:3857")
```

```python
# -----------------------------
# Step 5: Keep Required Columns Before Spatial Join
# -----------------------------
required_columns = ['geometry', 'slope', 'curve'] + all_dummy_cols
accidents_gdf = accidents_gdf[required_columns]

# Spatial join: assign accidents to grid cells
joined = gpd.sjoin(accidents_gdf, grid, how="left", predicate="within")
```

```python
# -----------------------------
# Step 6: Aggregate Per Grid Cell
# -----------------------------
# Count accidents per grid cell
counts_per_cell = joined.groupby('index_right').size().reset_index(name='accident_count')
grid['accident_count'] = 0
grid.loc[counts_per_cell['index_right'], 'accident_count'] = counts_per_cell['accident_count'].values

# Average slope, curve, and all dummy variables per cell
agg_all = joined.groupby('index_right')[['slope', 'curve'] + all_dummy_cols].mean().reset_index()

# Merge into grid
for col in agg_all.columns:
    if col != 'index_right':
        grid.loc[agg_all['index_right'], col] = agg_all[col].values
```

```python
# -----------------------------
# Step 7: Fit Poisson Regression Model
# -----------------------------
# Drop any cell with missing data
grid_model = grid.dropna(subset=['accident_count', 'slope', 'curve'])

# Features and target
features = ['slope', 'curve'] + all_dummy_cols
X = sm.add_constant(grid_model[features])
y = grid_model['accident_count']

# Fit the Poisson GLM
poisson_model = sm.GLM(y, X, family=sm.families.Poisson()).fit()

# Show results
print(poisson_model.summary())
```
